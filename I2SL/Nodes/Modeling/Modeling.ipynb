**LazyPredict**

LazyPredict helps us to predict and fit all Scikit learn models witha single line of code

This supports both Classification and Regression problems.
We will be performing both in this notebook.

In the inputs we will split our data into test and train datasets and use the LazyClassifier and the LazyRegressor in the output we get a dataframe which contains the performance summary and the predections of each model.

```python
from lazypredict.Supervised import LazyClassifier, LazyRegressor
from sklearn.model_selection import train_test_split
from sklearn import datasets
import pandas as pd
```

```python
#load data
data = datasets.load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=42)
```

**CLASSIFICATION**

```python
#fit models
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
#This data frame consists the performance summary of all models
models
```

```python
#This data frame contains all the predictions of each models
predictions
```

**REGRESSION**

```python
#load data
data = datasets.load_diabetes()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=42)
```

```python
#fit models
clf = LazyRegressor(predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
#This data frame consists the performance summary of all models
models
```

```python
#This data frame contains all the predictions of each models
predictions
```

**Conclusion**

LazyPredict is a convenient wrapper library that helps us quickly fit all the models into our dataset and this helps us compare there performances.
This helps us identitfy what works Out Of The Box.However this is not the best solution to model training and selection as only the default parameters are used.

Once we finalize which model we want to use we can start hyper parameter tuning of our models.
