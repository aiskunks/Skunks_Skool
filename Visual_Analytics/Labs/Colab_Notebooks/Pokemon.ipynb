```
num=3333
import random
def rnd_binary(b=11):
  r=bin(random.getrandbits(b)) #   getrandbits(k) -> x.  Generates a long int with k random bits.
  return r
```

```
def rnd_bits(b=11):
  r=random.getrandbits(b) #   getrandbits(k) -> x.  Generates a long int with k random bits.
  return r
```

```
def rnd_int(b=11):
  r=random.randint(0,2**b) #    random.randint(0,2**b)
  return r
```
